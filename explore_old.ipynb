```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import parallel_coordinates
from pandas.tools.plotting import radviz
```

```python
#Load data
test = pd.read_csv("data/csv/testset.csv", encoding = 'utf8')
train = pd.read_csv("data/csv/trainset.csv", encoding = 'utf8')
original = pd.read_csv("data/csv/original.csv", encoding = 'utf8')
# train.replace(["normal", "slowloris", "flooding"], [1, 2, 3], inplace = True)
# test.replace(["normal", "slowloris", "flooding"], [1, 2, 3], inplace = True)
# original.replace(["normal", "slowloris", "flooding"], [1, 2, 3], inplace = True)
```

```python
width = 12
height = 10
```

```python
def plot_class_bars(data, name):
    plt.figure()
    plt.suptitle("{} Dataset Class Percentage".format(name), fontsize=24)
    counts = data.xs("Class", axis=1).value_counts().astype(float).apply(lambda x: (x/data.shape[0]) * 100)
    counts.plot.bar(figsize=(width, height))
    plt.savefig("{}-counts.png".format(name.lower()))
```

```python
plot_class_bars(test, "Test")
```

```python
plot_class_bars(train, "Train")
```

```python
plot_class_bars(original, "Original")
```

```python
plt.figure(figsize=(width, height))
plt.suptitle("Test Dataset RadViz Plot", fontsize=24)
radviz(test, 'Class')
plt.savefig('test-radviz.png')
```

```python
plt.figure(figsize=(width, height))
plt.suptitle("Training Dataset RadViz Plot", fontsize=24)
radviz(train, 'Class')
plt.savefig('train-radviz.png')
```

```python
import pandas as pd
import scipy.stats as stats
import pylab 
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
%matplotlib inline 
```

```python
rawdata = pd.read_csv("../csv/expanded_dataset_v1.csv", encoding = 'utf8')
```

```python
norm_data = pd.read_csv("../csv/expanded_dataset_norm_v1.csv")
```

```python
scaler = MinMaxScaler()
minmax_data = scaler.fit_transform(rawdata.drop(['Class'], axis=1))
```

```python
scaled_data = preprocessing.scale(rawdata.drop(['Class'], axis=1))
```

```python
stats.probplot(rawdata.iloc[:,6], dist="norm", plot=pylab)
pylab.show()
```

```python
stats.probplot(norm_data.iloc[:,6], dist="norm", plot=pylab)
pylab.show()
```

```python
stats.probplot(minmax_data[:,6], dist="norm", plot=pylab)
pylab.show()
```

```python
stats.probplot(scaled_data[:,6], dist="norm", plot=pylab)
pylab.show()
```
